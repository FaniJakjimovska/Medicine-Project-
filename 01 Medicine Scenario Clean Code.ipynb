### 6. A normal hemoglobin A1C glucose level is below 5.7. Divide the data in 2 groups (patients that have normal A1C level and patients that have elevated A1C level), then sketch the data using a plot of your choice dependent on the length of the disease. Distinguishing both groups.

```python
df_hemoglobin_avg_duration = df.dropna(subset = (['length_of_the_condition', 'last_measured_value_of_the_hemoglobin_A1C_sugar']))
df_hemoglobin_avg_duration['A1C_level'] = df_hemoglobin_avg_duration['last_measured_value_of_the_hemoglobin_A1C_sugar']
df_hemoglobin_avg_duration['A1C_level'][df_hemoglobin_avg_duration['last_measured_value_of_the_hemoglobin_A1C_sugar'] < 5.7] = 'normal'
df_hemoglobin_avg_duration['A1C_level'][df_hemoglobin_avg_duration['last_measured_value_of_the_hemoglobin_A1C_sugar'] >= 5.7] = 'elevated'
df_hemoglobin_avg_duration_final = df_hemoglobin_avg_duration[['length_of_the_condition', 'A1C_level']]
df_hemoglobin_avg_duration_final.reset_index(inplace = True)
df_hemoglobin_avg_duration_final = df_hemoglobin_avg_duration_final.groupby(['length_of_the_condition', 'A1C_level']).count()
df_hemoglobin_avg_duration_final.reset_index(inplace=True)
df_hemoglobin_avg_duration_final = df_hemoglobin_avg_duration_final.rename(columns = {'index' : 'num_of_patients'})
df_hemoglobin_avg_duration_final.set_index(['length_of_the_condition', 'A1C_level'], inplace=True)
df_hemoglobin_avg_duration_final
```

```python
df_hemoglobin_avg_duration_final.reset_index(inplace=True)
group_labels = df_hemoglobin_avg_duration_final['A1C_level']


plt.plot(figsize = (20,5))



plt.bar( x = df_hemoglobin_avg_duration_final['length_of_the_condition'], height = df_hemoglobin_avg_duration_final['num_of_patients'], color = ['blue' if label == 'normal' else 'red' for label in group_labels], label=group_labels)

plt.legend(df_hemoglobin_avg_duration_final['A1C_level'].unique(), bbox_to_anchor=(1.25, 1), loc='upper right', borderaxespad=0)

plt.xlabel('Disease Length (years)')
plt.ylabel('Number of patients')
plt.title('Hemoglobin A1C vs. Disease Length')

plt.show()
```

```python
# Sample data
data = {
    'has(1)_or_does_not_have(0)_diabetic_retinopathy': [1.0, 1.0, 0.0, 0.0, 1.0],
    'insulin(1)_or_tablets(0)': [0.0, 0.0, 0.0, 1.0, 1.0],
    'last_measured_value_of_the_hemoglobin_A1C_sugar': [9.522692, 6.397994, 7.079202, 4.400294, 4.746816]
}

df_diabetic_retinopathy = pd.DataFrame(data)

# Grouping and counting
df_diabetic_retinopathy_grouped = df_diabetic_retinopathy.groupby(['has(1)_or_does_not_have(0)_diabetic_retinopathy', 'insulin(1)_or_tablets(0)']).size().unstack(fill_value=0)

# Plotting
fig, ax = plt.subplots(figsize=(10, 5))

# Bar plot
bar_width = 0.35
bar_positions = np.arange(len(df_diabetic_retinopathy_grouped))

# Bar for insulin(1)_or_tablets(0) == 0.0
ax.bar(bar_positions, df_diabetic_retinopathy_grouped[0.0], width=bar_width, label='Insulin(1)_or_tablets(0) = 0.0')

# Bar for insulin(1)_or_tablets(0) == 1.0
ax.bar(bar_positions + bar_width, df_diabetic_retinopathy_grouped[1.0], width=bar_width, label='Insulin(1)_or_tablets(0) = 1.0')

ax.set_xticks(bar_positions + bar_width / 2)
ax.set_xticklabels(df_diabetic_retinopathy_grouped.index)

ax.set_xlabel('has(1)_or_does_not_have(0)_diabetic_retinopathy')
ax.set_ylabel('Number of patients')
ax.set_title('Hemoglobin A1C vs. Diabetic Retinopathy and Insulin/Tablets')

ax.legend(title='Insulin(1)_or_tablets(0)')

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named 'df' with relevant columns
# df = ...

# Separating patients into two groups based on diabetic retinopathy
grouped_data = df.groupby('has(1)_or_does_not_have(0)_diabetic_retinopathy')

# Analyzing the characteristics of the groups
grouped_statistics = grouped_data.describe()

# Displaying statistics
print(grouped_statistics)

# Plotting various data points
# Example: Box plots for numerical variables
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))

sns.boxplot(x='has(1)_or_does_not_have(0)_diabetic_retinopathy', y='length_of_the_condition', data=df, ax=axes[0, 0])
axes[0, 0].set_title('Length of the Condition')

sns.boxplot(x='has(1)_or_does_not_have(0)_diabetic_retinopathy', y='last_measured_value_of_the_hemoglobin_A1C_sugar', data=df, ax=axes[0, 1])
axes[0, 1].set_title('Hemoglobin A1C Level')

sns.boxplot(x='has(1)_or_does_not_have(0)_diabetic_retinopathy', y='insulin(1)_or_tablets(0)', data=df, ax=axes[1, 0])
axes[1, 0].set_title('Insulin or Tablets')

sns.boxplot(x='has(1)_or_does_not_have(0)_diabetic_retinopathy', y='current_glucose_level', data=df, ax=axes[1, 1])
axes[1, 1].set_title('Current Glucose Level')

plt.tight_layout()
plt.show()

# Correlation matrix
correlation_matrix = df.corr()

# Displaying correlation matrix
print(correlation_matrix)

# Heatmap of the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```
